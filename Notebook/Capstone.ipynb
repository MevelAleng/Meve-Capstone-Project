```python
cdf = dfc.iloc[countries_code].copy()
```

```python
cdf
```

```python
cdf.columns = cdf.columns.str.replace('!!Estimate', '')
```

```python
cdf.rename(columns={cdf.columns[0]: 'Country'}, inplace=True)
```

```python
cdf['Country'] = cdf['Country'].str.strip()
```

```python
cdf_clean = cdf.melt(id_vars=['Country'], var_name='State', value_name='Population')
```

```python
cdf_clean['Population'] = (
    cdf_clean['Population']
    .astype(str)
    .str.replace(',', '', regex=False)    
    .pipe(pd.to_numeric, errors='coerce') 
    .fillna(0)                            
    .astype(int)                          
)
```

```python
cdf_fin = cdf_clean[cdf_clean['State'] != 'United States']
```

```python
cdf_fin = cdf_fin.copy()
```

```python
countrytots = cdf_fin.groupby('Country')['Population'].transform('sum')
```

```python
cdf_fin['Country Total %'] = (cdf_fin['Population'] / countrytots) * 100
```

```python
statetots = cdf_fin.groupby('State')['Population'].transform('sum')
```

```python
cdf_fin['State Total %'] = (cdf_fin['Population'] / statetots) * 100
```

```python
cdf_fin
```

```python
cdf_fin.groupby('State')['State Total %'].sum()
```

```python
cdf_fin.to_csv('cdfperc.csv', index=False)
```

Total comparison of them genders for sudan and south sudan in the year 2024.

```python
gendertotal = egpytasylum.groupby('Country of Origin').agg({'Female Total': 'sum', 'Male Total': 'sum', 'Total': 'sum'})
```

```python
gendertotal
```

```python
gendertotal['Children'] = egpytasylum.groupby('Country of Origin')[childgroups].sum().sum(axis=1)
```

```python
gendertotal['Adult'] = egpytasylum.groupby('Country of Origin')[adultgroups].sum().sum(axis=1)
```

```python
gendertotal
```

Percantage of the gender and age totals.

```python
gendertotal['Child %'] = (gendertotal['Children'] / gendertotal['Total']) * 100
```

```python
gendertotal['Adult %'] = (gendertotal['Adult'] / gendertotal['Total']) * 100
```

```python
gendertotal['Male %'] = (gendertotal['Male Total'] / gendertotal['Total']) * 100
```

```python
gendertotal['Female %'] = (gendertotal['Female Total'] / gendertotal['Total']) * 100
```
